```python
# Sample data (replace this with your actual dataset)
data = pd.DataFrame({
    'country': ['Albania', 'Albania', 'Albania', 'Albania', 'Albania'],
    'year': [1987, 1987, 1987, 1987, 1987],
    'sex': ['male', 'male', 'female', 'male', 'male'],
    'age': ['15-24 years', '35-54 years', '15-24 years', '75+ years', '25-34 years'],
    'generation': ['Generation X', 'Silent', 'Generation X', 'G.I. Generation', 'Boomers']
})

# Create a bar chart using seaborn countplot
plt.figure(figsize=(10, 6))
sns.countplot(x='generation', data=data, palette='viridis')
plt.xlabel('Generation')
plt.ylabel('Count')
plt.title('Count of Generations')
plt.xticks(rotation=45)
plt.show()
```

```python


# Sample data (replace this with your actual dataset)
data = pd.DataFrame({
    'country': ['Albania', 'Albania', 'Albania', 'Albania', 'Albania'],
    'year': [1987, 1987, 1987, 1987, 1987],
    'sex': ['male', 'male', 'female', 'male', 'male'],
    'age': ['15-24 years', '35-54 years', '15-24 years', '75+ years', '25-34 years'],
    'generation': ['Generation X', 'Silent', 'Generation X', 'G.I. Generation', 'Boomers']
})

# Group the data by 'generation' and count occurrences
generation_counts = data['generation'].value_counts()

# Calculate explode list dynamically based on the number of parts
num_parts = len(generation_counts)
explode = [0.1 if i == generation_counts.idxmax() else 0 for i in range(num_parts)]

# Create a pie chart with some distance between parts, shadows, and different colors using matplotlib
plt.figure(figsize=(8, 8))
colors = ['skyblue', 'lightcoral', 'gold', 'lightgreen', 'lightpink']  # List of colors for each part
plt.pie(generation_counts, labels=generation_counts.index, autopct='%1.1f%%', startangle=140, colors=colors, explode=explode, )
plt.axis('equal')  # Equal aspect ratio ensures that the pie is drawn as a circle.
plt.title('Distribution of Generations')
plt.show()
```
